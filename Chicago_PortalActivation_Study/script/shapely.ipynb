```python
from shapely.geometry import shape
import json

with open('zipcodes.geojson') as f:
    data = json.load(f)

for feature in data['features']:
    zipcode = feature['properties']['zip']
    geometry = shape(feature['geometry'])
    center_long, center_lat = geometry.centroid.xy
    print(f"Zipcode {zipcode}: ({center_long[0]}, {center_lat[0]})")
```
